#### Корреляция и косинусное сходство

##### **Корреляция**  
Корреляция измеряет степень линейной зависимости между двумя переменными. В Data Science чаще всего используется **коэффициент корреляции Пирсона**:  

\$
r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2} \cdot \sqrt{\sum (y_i - \bar{y})^2}}
\$

Где:  
- \$ x_i \$ и \$ y_i \$ — элементы векторов,  
- \$ \bar{x} \$ и \$ \bar{y} \$ — средние значения соответствующих векторов.  

- \$ r \$ принимает значения от -1 до 1:
  - \$ r \approx 1 \$ – сильная прямая связь (одна переменная растёт → другая тоже).
  - \$ r \approx -1 \$ – сильная обратная связь (одна растёт → другая уменьшается).
  - \$ r \approx 0 \$ – связи нет.  



**Применение в Data Science:**  
- **Анализ данных:** Определение взаимосвязей между признаками (feature selection).  
- **Финансовый анализ:** Оценка зависимости акций, макроэкономических показателей.  
- **Медицина:** Связь между симптомами и болезнями.  

---

##### **Косинусное сходство**  
Косинусное сходство измеряет угол между двумя векторами в многомерном пространстве:

\$
\cos(\theta) = \frac{A \cdot B}{||A|| \cdot ||B||} = \frac{\sum A_i B_i}{\sqrt{\sum A_i^2} \cdot \sqrt{\sum B_i^2}}
\$
Где:  
- \$ A \$ и \$ B \$ — два вектора.  
- \$ A \cdot B \$ — **скалярное произведение** векторов, вычисляется как сумма попарных произведений элементов.  
- \$ ||A|| \$ — **длина (норма) вектора A**, вычисляется как корень из суммы квадратов элементов.  
- \$ ||B|| \$ — **длина (норма) вектора B**.  

##### **Что значат значения косинусного сходства?**
- Принимает значения от -1 до 1:
  - \$ 1 \$ – векторы полностью совпадают (имеют одинаковое направление).
  - \$ 0 \$ – векторы ортогональны (перпендикулярны, не имеют ничего общего).
  - \$ -1 \$ – векторы полностью противоположны.  

**Применение в Data Science:**  
- **Рекомендательные системы:** Сравнение профилей пользователей или товаров (например, фильмы на Netflix).  
- **Обработка текстов (NLP):** Определение схожести документов (bag-of-words, TF-IDF).  
- **Поиск дубликатов:** Поиск похожих изображений, аудиофайлов, статей.  

##### **Ключевые различия**  
| Метрика | Когда использовать? |  
|---------|----------------------|  
| **Корреляция** | Если важна линейная зависимость и масштаб данных (например, температура и спрос на мороженое). |  
| **Косинусное сходство** | Если важен угол между векторами, а не их длины (например, схожесть текстов независимо от количества слов). |  

Если работаешь с текстами, рекомендательными системами или анализом многомерных данных – косинусное сходство твой инструмент. Если ищешь линейные связи в данных – корреляция.

#### Фильтрация временных рядов и обнаружение признаков

#### **1. Фильтрация временных рядов**  
Фильтрация используется для устранения шума, выявления трендов и сезонных компонентов во временных рядах.  

##### **Скользящее среднее (Moving Average, MA)**  
Один из самых простых методов фильтрации:  

\$
\tilde{x_t} = \frac{1}{N} \sum_{i=0}^{N-1} x_{t-i}
\$

Где:  
- \$ \tilde{x_t} \$ – сглаженное значение,  
- \$ N \$ – размер окна сглаживания,  
- \$ x_t \$ – исходные значения временного ряда.



**Применение в Data Science:**  
- Сглаживание временных рядов перед моделированием.  
- Удаление случайного шума из данных.  
- Определение трендов и сезонов.  

---

##### **Экспоненциальное сглаживание (Exponential Smoothing, EWMA)**  
Придает больше веса недавним наблюдениям:  

\$
\tilde{x_t} = \alpha x_t + (1 - \alpha) \tilde{x_{t-1}}
\$

Где:  
- \$ \alpha \in (0, 1) \$ – коэффициент сглаживания.  

**Применение:**  
- Финансовые рынки (анализ котировок акций).  
- Предсказание временных рядов (прогнозирование спроса).  

---

##### **Фильтр Калмана (Kalman Filter)**  
Динамический метод, учитывающий скрытые переменные и шум в данных. Используется в машинном обучении и предсказательном анализе.  

\$
x_t = A x_{t-1} + w_t
\$
\$
z_t = H x_t + v_t
\$

Где:  
- \$ x_t \$ – скрытое состояние,  
- \$ z_t \$ – наблюдение,  
- \$ A, H \$ – матрицы системы,  
- \$ w_t, v_t \$ – шумы.  

**Применение:**  
- Отслеживание объектов (например, GPS-трекинг).  
- Фильтрация данных IoT (интернет вещей).  

---

#### **2. Обнаружение признаков во временных рядах**  
Обнаружение признаков (features) – важная часть Data Science, позволяющая выделять значимые паттерны из временных рядов.  

##### **Выявление аномалий (Anomaly Detection)**  
Используются статистические и ML-методы:  
- **Z-score**:  
  \$
  Z_t = \frac{x_t - \mu}{\sigma}
  \$ 
  Если \$ |Z_t| > 3 \$ → выброс.  

- **Метод локтя (DBSCAN, Isolation Forest)** – для обнаружения выбросов в данных.  

**Применение:**  
- Обнаружение мошенничества в транзакциях.  
- Мониторинг сетевой безопасности.  
- Выявление сбоев оборудования.  

---

##### **Автокорреляция (Autocorrelation Function, ACF)**  
Измеряет зависимость ряда от его прошлых значений:  

\$
ACF(k) = \frac{\sum (x_t - \mu) (x_{t-k} - \mu)}{\sum (x_t - \mu)^2}
\$

Используется для анализа сезонности.  

**Применение:**  
- Прогнозирование спроса в ритейле.  
- Анализ поведения пользователей (например, циклы активности).  

---

##### **Фурье-анализ (FFT – Fast Fourier Transform)**  
Применяется для анализа частотных составляющих временного ряда.  

\$
X(f) = \sum x_t e^{-j 2 \pi f t}
\$

Позволяет находить циклы, сезонные тренды.  

**Применение:**  
- Анализ сигналов (аудио, электроэнцефалограмма).  
- Оптимизация бизнес-процессов (анализ производственных данных).  

---

#### **Вывод**  
Фильтрация временных рядов помогает убирать шум и выявлять тренды, а обнаружение признаков позволяет находить аномалии и сезонные паттерны. В Data Science это критически важно для финансов, IoT, маркетинга и других областей.

#### Кластеризация методом k-средних

#### **1. Суть метода k-средних**  
Алгоритм делит набор данных на \$ k \$ кластеров, минимизируя внутрикластерное расстояние (инерцию).  

##### **Формальная постановка задачи**  
Для множества точек данных \$ X = \{x_1, x_2, ..., x_n\} \$ находим \$ k \$ центроидов \$ C = \{c_1, c_2, ..., c_k\} \$, минимизируя сумму квадратов расстояний от точек до ближайшего центроида:

\$
J = \sum_{i=1}^{k} \sum_{x \in S_i} || x - c_i ||^2
\$

Где:  
- \$ S_i \$ – множество точек в i-м кластере,  
- \$ c_i \$ – центроид i-го кластера,  
- \$ || x - c_i ||^2 \$ – евклидово расстояние.  

##### **Шаги алгоритма**  
1. **Инициализация**: выбираем \$ k \$ случайных центроидов.  
2. **Назначение точек**: каждая точка \$ x \$ относится к ближайшему центроиду по евклидову расстоянию.  
3. **Пересчет центроидов**: новый центроид – это среднее значение всех точек в кластере:  

   \$
   c_i = \frac{1}{|S_i|} \sum_{x \in S_i} x
   \$

4. **Повторение шагов 2-3** до сходимости (когда центроиды перестают изменяться).  

---

#### **2. Применение в Data Science**  

##### **1️⃣ Сегментация клиентов**  
Используется в маркетинге и ритейле:  
- Группировка покупателей по поведению.  
- Персонализированные рекомендации.  
- Анализ оттока клиентов.  

##### **2️⃣ Обнаружение аномалий**  
Обнаружение **выбросов** (аномалий) в данных, например:  
- Мошеннические транзакции в банке.  
- Ошибки датчиков IoT.  
- Подозрительное поведение пользователей.  

##### **3️⃣ Сжатие изображений (Image Compression)**  
Метод \$ k \$-средних используется для **кластеризации цветов** в изображении.  
- Уменьшение количества цветов (разбиение пикселей на кластеры).  
- Оптимизация хранения и передачи данных.  

##### **4️⃣ Биология и медицина**  
- Анализ ДНК (группировка схожих последовательностей).  
- Кластеризация опухолевых клеток.  
- Разделение пациентов по группам риска.  

---

#### **3. Ограничения метода k-средних**  
**Неоптимальный выбор \$ k \$**: нужно подбирать количество кластеров. Используют **метод локтя**:  

\$
WSS(k) = \sum_{i=1}^{k} \sum_{x \in S_i} || x - c_i ||^2
\$

Где \$ WSS(k) \$ – сумма внутрикластерных расстояний. Строят график \$ WSS \$ и выбирают точку перегиба ("локоть").  

**Чувствительность к выбросам**: сильные выбросы могут сильно смещать центроиды.  

**Форма кластеров**: алгоритм плохо работает, если кластеры не имеют **сферической формы**. В таких случаях лучше использовать **DBSCAN или иерархическую кластеризацию**.  

---

#### **Вывод**  
Метод \$ k \$-средних – мощный инструмент для кластеризации данных в Data Science, используемый в маркетинге, медицине, анализе аномалий и сжатии изображений. Однако он требует выбора \( k \), чувствителен к выбросам и плохо работает с некруглыми кластерами.

Коэффициент корреляции Пирсона
\$
r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2} \cdot \sqrt{\sum (y_i - \bar{y})^2}}
\$

Где:  
- \$ x_i \$ и \$ y_i \$ — элементы векторов,  
- \$ \bar{x} \$ и \$ \bar{y} \$ — средние значения соответствующих векторов.  

```python
import numpy as np
import scipy
import matplotlib.pyplot as plt

import time
```

```python
vec1 = np.array([1,2,3,4])
vec2 = np.array([4,-2,2,1])
```

```python
print(np.corrcoef(vec1,vec2))
```

```python
def PirsonCorr(vec1, vec2):
    # проверка размерности векторов
    if len(vec1) != len(vec2):
        raise ValueError('Векторы должны быть одной размерности')

    n = len(vec1)
    mean_x = sum(vec1) / n
    mean_y = sum(vec2) / n

    # вычисление числителя
    numerator = sum((x - mean_x) * (y - mean_y) for x, y in zip(vec1, vec2))
    # вычисление части знаменателя от x
    denominator_x = sum((x-mean_x) ** 2 for x in vec1) ** 0.5
    # вычисление части знаменателя от y
    denominator_y = sum((y-mean_y) ** 2 for y in vec2) ** 0.5

    if denominator_x == 0 or denominator_y == 0:
        return 0 # Корреляция не определена, если один из векторов - константа
    # результат
    result = numerator / (denominator_x * denominator_y)
    return result
```

```python
print(PirsonCorr(vec1, vec2))
```

Косинусное сходство измеряет угол между двумя векторами в многомерном пространстве:

\$
\cos(\theta) = \frac{A \cdot B}{||A|| \cdot ||B||} = \frac{\sum A_i B_i}{\sqrt{\sum A_i^2} \cdot \sqrt{\sum B_i^2}}
\$

- \$ A \$ и \$ B \$ — два вектора.  
- \$ A \cdot B \$ — **скалярное произведение** векторов, вычисляется как сумма попарных произведений элементов.  
- \$ ||A|| \$ — **длина (норма) вектора A**, вычисляется как корень из суммы квадратов элементов.  
- \$ ||B|| \$ — **длина (норма) вектора B**.  

```python
def CosineSimilarity(vec1, vec2):
    # проверка размерности векторов
    if len(vec1) != len(vec2):
        raise ValueError('Векторы должны быть одной размерности')
        
    # Вычисляем скалярное произведение
    numerator = np.dot(vec1,vec2)
    # Вычисляем нормы (длины) векторов
    vec_norm_1 = (sum([x**2 for x in vec1])) ** 0.5
    vec_norm_2 = (sum([x**2 for x in vec2])) ** 0.5
    
    # Проверка деления на ноль (если один из векторов нулевой)
    if vec_norm_1 == 0 or vec_norm_2 == 0:
        return 0  # Если хотя бы один вектор нулевой, сходство неопределено, возвращаем 0

    result = numerator / (vec_norm_1 * vec_norm_2)

    return result
```

```python
print(CosineSimilarity(vec1, vec2))
```

```python
# функция из предыдущего упражнения, как ее видит автор...

def corrAndCosine(x,y):

  # compute cosine similarity
  num = np.dot(x,y) # numerator
  den = np.linalg.norm(x) * np.linalg.norm(y) # denominator
  cos = num / den

  # compute correlation (similar to above but mean-centered!)
  xm  = x-np.mean(x)
  ym  = y-np.mean(y)
  num = np.dot(xm,ym) # numerator
  den = np.linalg.norm(xm) * np.linalg.norm(ym) # denominator
  cor = num / den

  return cor,cos
```

```python
# создание переменной
a = np.arange(4,dtype=float)
# создание смещения
offsets = np.arange(-50,51)

# инициализация результатов
results = np.zeros((len(offsets),2))

# симуляция в цикле
# получаем корреляцию и косинусное сходство
for i in range(len(offsets)):
    results[i,:] = corrAndCosine(a,a+offsets[i])


# plot the results!
plt.figure(figsize=(8,4))
h = plt.plot(offsets,results)
h[0].set_color('k')
h[0].set_marker('o')
h[1].set_color([.7,.7,.7])
h[1].set_marker('s')

plt.xlabel('Mean offset')
plt.ylabel('r or c')
plt.legend(['Pearson','Cosine sim.'])
plt.savefig('Figure_04_02.png',dpi=300) # write out the fig to a file
plt.show()
```
